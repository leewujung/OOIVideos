#### Visualizing bounds of static scenes from videos taken on 2016/01/01

Videos are subsampled at rate 10.

```python
video_path = os.path.join(os.getcwd(),'..','results','bounds_20160101')
```

```python
listOfFiles = sorted(os.listdir(video_path))
```

```python
fig = plt.figure(figsize = (20,5))
height = 1
fig.canvas.draw()
for item in listOfFiles:
    if '0101' in item:
        bounds = pd.read_csv(os.path.join(video_path,item))
        plt.plot(bounds['LB'],height*np.ones(len(bounds['LB'])),'ro')
        plt.plot(bounds['UB'],height*np.ones(len(bounds['LB'])),'bo')
        height = height + 1
        plt.ylim([0,10])
        plt.text(-300, height-1, item[7:-5])
        
plt.axes().get_yaxis().set_visible(False)
plt.title('20160101 Scene Bounds'  )

```

Beginning of a scene is a red dot and end of scene is a blue dot. The seem to get misaligned toward the end of the video.

#### Visualizing bounds of static scenes from videos taken across different weeks

```python
video_path = os.path.join(os.getcwd(),'..','results','bounds_weekly')
```

```python
listOfFiles = sorted(os.listdir(video_path))
```

```python
fig = plt.figure(figsize = (20,5))
height = 1
fig.canvas.draw()
for item in listOfFiles:
    if '.csv' in item:
        bounds = pd.read_csv(os.path.join(video_path,item))
        plt.plot(bounds['LB'],height*np.ones(len(bounds['LB'])),'ro')
        plt.plot(bounds['UB'],height*np.ones(len(bounds['LB'])),'bo')
        height = height + 1
        plt.ylim([0,10])
        plt.text(-300, height-1, item[7:-5])
    
plt.axes().get_yaxis().set_visible(False)

plt.title('Weekly Scene Bounds'  )
```

```python
video_path = os.path.join(os.getcwd(),'..','results','rolling_variance_weekly')
```

```python
listOfFiles = sorted(os.listdir(video_path))
```

```python
fig = plt.figure(figsize = (20,5))
height = 1

for item in listOfFiles:
    
    if '.csv' in item:
        var = pd.read_csv(os.path.join(video_path,item))
        plt.plot(var+height*2e9)
   
    plt.text(-300, height*2e9, item[16:-5])
    height = height + 1
plt.axes().get_yaxis().set_visible(False)
    
plt.title('Weekly Rolling Variance'  )
```
